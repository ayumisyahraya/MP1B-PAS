```python
import numpy as np

X_train = np.array(heart_attack_df[['Physical Activity Days Per Week', 'Sleep Hours Per Day']])
y_train = np.array(heart_attack_df['Heart Attack Risk'])

print(f'X_train:\n{X_train}\n')
print(f'y_train: {y_train}')

#Kode ini bertujuan untuk menyiapkan data latih (training data) dalam format NumPy array dari DataFrame heart_attack_df
#agar bisa digunakan untuk pembelajaran mesin (machine learning).
#Pembelajaran mesin memerlukan data dalam format numerik seperti array agar model bisa menghitung dan membuat prediksi.
```

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
print(f'y_train:\n{y_train}')
#Kode ini digunakan untuk mengubah label kategorikal pada y_train (seperti 'High', 'Low', dan 'Medium')
#menjadi format numerik menggunakan LabelBinarizer dari scikit-learn. Hal ini diperlukan ketika
#model pembelajaran mesin memerlukan input numerik (seperti dalam kasus klasifikasi).
```

```python
y_train = y_train.flatten()
print(f'y_train: {y_train}')

#Kode ini digunakan untuk mengubah bentuk y_train yang sebelumnya merupakan array dua dimensi menjadi array satu dimensi menggunakan fungsi flatten().
```

```python
from sklearn.neighbors import KNeighborsClassifier

K = 3
model = KNeighborsClassifier(n_neighbors=K)
model.fit(X_train, y_train)

#Kode ini digunakan untuk melatih model K-Nearest Neighbors (KNN)
#dengan 3 tetangga terdekat (K=3) menggunakan data pelatihan X_train (fitur) dan y_train (label).
```

```python
Physical_Activity_Days_Per_Week = 0
Sleep_Hours_Per_Day = 1
X_new = np.array([Physical_Activity_Days_Per_Week, Sleep_Hours_Per_Day]).reshape(1, -1)
X_new

#Kode ini digunakan untuk melatih model K-Nearest Neighbors (KNN) dengan 3 tetangga terdekat (K=3)
#menggunakan data pelatihan X_train (fitur) dan y_train (label).
```

```python
y_new = model.predict(X_new)
y_new

#Kode ini digunakan untuk melakukan prediksi dengan model K-Nearest Neighbors (KNN)
#yang telah dilatih sebelumnya, menggunakan data input baru (X_new), dan mengeluarkan prediksi untuk label output (y_new).
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
print (f'Accuracy: {acc}')

#Kode accuracy_score(y_test, y_pred) menghitung akurasi model dengan membandingkan label yang sebenarnya (y_test)
#dan label yang diprediksi (y_pred). Hasilnya, acc menunjukkan persentase prediksi yang benar.
```

```python
from sklearn.metrics import precision_score
prec = precision_score(y_test, y_pred)
print(f'Precission: {prec}')

#menghitung precision, yaitu seberapa akurat prediksi positif model.
```

```python
from sklearn.metrics import recall_score
rec = recall_score (y_test, y_pred)
print(f'Recall: {rec}')

#menghitung recall, yaitu seberapa banyak prediksi positif yang benar dibandingkan dengan semua data positif yang sebenarnya.
```

```python
from sklearn.metrics import f1_score
f1 = f1_score (y_test, y_pred)
print (f'F1-score: {f1}')

#enghitung F1-score, yang merupakan rata-rata harmonis antara precision dan recall, memberikan gambaran tentang keseimbangan antara keduanya.
```

```python
from sklearn.metrics import classification_report
cls_report = classification_report(y_test, y_pred)
print(f'Classification Report:\n{cls_report}')

#menghasilkan laporan yang mencakup metrik evaluasi seperti precision, recall, F1-score, dan support untuk setiap kelas, memberikan gambaran keseluruhan performa model.
```
